```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
#Veri setinde sütün isimleri olmadığından sütün isimlerini tanımlayalım

columns = [
     'state',
     'account length', 
     'area code', 
     'phone number', 
     'international plan', 
     'voice mail plan', 
     'number vmail messages',
     'total day minutes',
     'total day calls',
     'total day charge',
     'total eve minutes',
     'total eve calls',
     'total eve charge',
     'total night minutes',
     'total night calls',
     'total night charge',
     'total intl minutes',
     'total intl calls',
     'total intl charge',
     'number customer service calls',
     'churn']
```

```python
data = pd.read_csv('ChurnDataset.txt', header = None, names = columns)
data
```

```python
data.shape 
```

```python
siniflandirma={'no': 0., 'yes':1., 'False.':0., 'True.':1.}
data.replace({'international plan' : siniflandirma, 'voice mail plan' : siniflandirma, 'churn' : siniflandirma}, 
             regex = True, inplace = True)  #burada karakter değiştirme işlemini uyguladık
```

```python
data.isnull().sum()
```

```python
#Veri setini temizleyelim işimize yaramayan verileri kaldıralım
data.drop('phone number', axis = 1, inplace = True)
data.drop('area code', axis = 1, inplace = True)
data.drop('state', axis = 1, inplace = True)
data
```

```python
#Üçüncü ön işleme adımında ise “churn olan” ve “churn olmayan” örnek veri sayısını eşitlemeye çalışıyoruz. 
#Amacımız bu tarz veri setlerindeki en büyük problemlerden birisi olan 
#“sınıf dengesizliği” (class imbalance) problemini çözmek.

data1 = data[data['churn']==1]  #1 yani churn olanlar
data2 = data[data['churn']==0]  #0 yani churn olmayanlar
```

```python
data1.shape
```

```python
data2.shape
```

```python
#Sınıf dengesizliği problemini çözmek için 
#basit ama etkili bir çözüm uygulayalım: churn olmayan 483 tane örnek veriyi churn olan 483 tane örnek veri ile aşağıdaki gibi birleştirelim 
#ve ekrana son veri seti ile ilgili bilgi yazdıralım

yeniData=data = data1.append(data2[:483])
yeniData.shape
```

# Eğitim ve Test Verisinin Oluşturulması

```python
#Egitim  ve test verisini parcaliyoruz  80% / 20%
X = yeniData.iloc[:, yeniData.columns != 'churn']
y = yeniData['churn']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
model=MinMaxScaler()
model.fit(X_train,y_train)

#Datamızın özelliklerini/değişkenlerini normalleştirmenin 
#bir başka yolu (özellikleri μ=0 ve σ=1 olacak şekilde ölçekleyen 
#standardizasyon dışında) Min-Max ölçekleyicidir
```

```python
XX_train = model.transform(X_train.values)
XX_test  = model.transform(X_test.values)
```

```python
YY_train = y_train.values 
YY_test  = y_test.values
```

# Makine Öğrenimi Algoritmaları ile Model Değerlendirme


```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, BaggingClassifier
```

```python
#Daha sonra modelleri dinamik bir listeye ekliyoruz.
#Aşağıda da gördüğümüz üzere; listenin ilk elemanını “sınıflandırma modelinin ismi”, ikinci elemanını 
#“modelin kurucu (constructor) metodunun çağrılması” şeklinde tanımladık

models = []
models.append(('Logistic Regression', LogisticRegression()))
models.append(('Naive Bayes', GaussianNB()))
models.append(('Decision Tree (CART)',DecisionTreeClassifier())) 
models.append(('K-NN', KNeighborsClassifier()))
models.append(('SVM', SVC()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('AdaBoostClassifier', AdaBoostClassifier()))
models.append(('BaggingClassifier', BaggingClassifier()))
models.append(('RandomForestClassifier', RandomForestClassifier()))
```

```python
#Son adımda; her modeli “for döngüsü” kullanarak daha önce oluşturduğumuz 
#eğitim ve test veri setleri üzeriden eğitip, her modelin ACC (Accuracy / Doğruluk) 
#ölçütüne göre başarısını elde ediyoruz


# Modelleri test edelim
for name, model in models:
    model = model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    from sklearn import metrics
    print("Model -> %s -> ACC: %%%.2f" % (name,metrics.accuracy_score(y_test, y_pred)*100))
```
